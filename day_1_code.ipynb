```python
with open("day_1.txt") as fin:
    lines = fin.read().strip().split("\n")
```

#### Part 1

```python
location = 50 
ans=0
for i in lines: 
    rotation = i[0]
    val = int(i[1:])
    if rotation == "L": 
        location = (location - val +100)%100
    else: 
        location = (location + val)%100
    
    if location == 0: 
        ans+=1
        
print(ans)  
```

#### Part 2

```python
location = 50 
ans=0
for i in lines: 
    rotation = i[0]
    val = int(i[1:])
    
    for _ in range(val): 
        if rotation == "L": 
            location = (location - val +100)%100
        else: 
            location = (location + val)%100
    
        if location == 0: 
            ans+=1
        
print(ans)  
```
