```python
with open("day_2.txt") as fin:
    lines = fin.read().strip().split(",")
ans=0       
```

#### Part 1

```python
for i in lines: 
    f,l = i.split("-")
    l,f=int(l), int(f)
    for j in range(f, l+1): 
        j=str(j)
        if len(j)%2==0 and j[:len(j)//2]==j[len(j)//2:]: 
            ans+=int(j)
print(ans)
```
